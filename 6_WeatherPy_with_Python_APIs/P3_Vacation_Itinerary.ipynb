```python
#### HOW TO ADD THIS LAYER (MARKER LAYER) INTO THE MAP BELOW WITH THE DIRECTIONS

# Set up marker layer for #10 in part 3
c_s = cities.copy()
c_s['Location'] = [(Lat, Lng) for Lat, Lng in zip(c_s.Lat, c_s.Lng)]
c_s = c_s.to_dict('r')

city_locations = [c['Location'] for c in c_s]
info_box_template = """
<dl>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Hotel</dt><dd>{Hotel}</dd>
<dt>Max Temp</dt><dd>{Max_Temp}</dd>
<dt>Humidity</dt><dd>{Humidity}</dd>
<dt>Cloudiness</dt><dd>{Cloudiness}</dd>
<dt>Rain</dt><dd>{Rain_Inches}</dd>
<dt>Snow</dt><dd>{Snow_Inches}</dd>
</dl>
"""

city_info = [info_box_template.format(**c) for c in c_s]
```
