```python
file = pd.read_csv('WeatherPy_database.csv')
file = file[['City', 'Country', 'Max Temp', 'Humidity', 
             'Cloudiness', 'Rain (Inches)', 'Snow (Inches)', 
             'Wind Speed', 'Lat', 'Lng', 'Date']]
```
